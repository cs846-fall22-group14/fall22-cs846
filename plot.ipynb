```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
allLines = []
with open("num_used_features.txt") as file:
    allLines = file.readlines()
```

```python
allNums = []
for l in allLines:
    allNums.append(int(l))
```

```python
counts, bins = np.histogram(allNums, bins=range(13))
plt.title("features per line")
plt.ylabel("programs")
plt.hist(bins[:-1], bins, weights=counts, align="left")
```
